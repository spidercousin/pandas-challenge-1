## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

df['Total Price Before'] = df['unit_price'] * df['qty']
df['Total Weight'] = df['qty'] * df['unit_weight']
# df['unit_price'] = df['unit_price']
# df['Total Price Before'] = df['Total Price Before']
# df['unit_cost'] = df['unit_cost']
minus_columns = df
minus_columns.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds
# $10 per pound for items 50 pounds or under.
def calculate_shipping(row):
    if row['Total Weight'] > 50:
        return 7 * row['Total Weight']
    else:
        return 10 * row['Total Weight']



minus_columns['Shipping Cost'] = minus_columns.apply(calculate_shipping, axis=1)
minus_columns['Shipping Cost'] = minus_columns['Shipping Cost'].round(2)
# shipping = minus_columns.loc[minus_columns['unit_weight'] > 50, ['unit_weight', 'Shipping Cost']]


shipping = minus_columns
shipping.head()











```

```python


# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

# def shippingtax(row):
#     row['Total Price After'] = row['Total Price Before'] + (row['Total Price Before'] * 0.0925) + row['Shipping Price']
#     return row

# tax_price = shipping['Total Price Before'] * 0.925
# shipping['Total Price After'] = shipping['Total Price Before'] + shipping['Shipping Price'] + tax_price
# shipping

tax_rate = 1.0925
# shipping['Total Price Before'] = shipping['Total Price Before'] * tax_rate
shipping['Total Price After Tax'] = (shipping['Total Price Before'] + shipping['Shipping Cost']) * tax_rate

shipping.round(2)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

shipping['Order Cost'] = shipping['unit_cost'] * shipping['qty'] + shipping['Shipping Cost']
shipping.round(2)

```

```python
# Create a column for the profit of each line using line cost and line price
shipping['Profit'] = shipping['Total Price After Tax'] - shipping['Order Cost']


shipping.iloc[:, 7:24].round(2)

```
